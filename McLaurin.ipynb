##Aproxiamcion de MacLaurin
Calcular la serie de McLaurin de sexto orden para $$f(x)=\cos(x)$$

**Solucion:** La serie de McLaurin es $$\sum_{i=0}^n \frac{f^{(i)}(x)}{i!}x^i$$

Por lo que $$f(x)≈1-\frac{x^2}{2}+\frac{x^4}{24}-\frac{x^5}{720}$$

```python
import matplotlib.pyplot as plt
import numpy as np

#Crea un espacio lineal en el intervalo -5 y 5 y con 100 puntos entre los valores numericos
x = np.linspace(-5,5, 100)
#Funcion coseno
y = np.cos(x)
#Polinomio de MacLaurin de orden 6 del coseno centrado en 0
y_approx = 1 - (x**2 / 2) + (x**4 / 24) - (x**6 / 720)

#Crea la figura
plt.figure(figsize=(10, 10))
#Crea la grafica dxe la funcion coseno
plt.plot(x, y, label="cos(x)", color='blue')
#Crea la grafica de la aproximacion dandole u estilo punteado y de color naranja
plt.plot(x, y_approx, label="Aproximacion de McLaurin de orden 6", color='orange', linestyle='--')
#Muestra los ejes x y y
plt.axvline(x=0, color='black', linestyle='--')
plt.axhline(y=0, color='black', linestyle='--')

#Agrega titulo, etiquetas a ejes , viñetas y cuadriculado
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

De esta podemos observar que en un intervalo el polinomio coincide al $\cos(x)$
