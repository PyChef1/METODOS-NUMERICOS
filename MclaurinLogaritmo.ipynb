#Serie de McLaurin

Anteriormente sabiamos que la serie de McLaurin es: $$\sum_{i=0}^n \frac{f^{(i)}(x)}{i!}x^i$$

Por lo que si $f(x)= \ln{x}$ la aproximacion de McLaurin sera: $$f(x)\approx x+\frac{(x-1)^3}{3}-\frac{(x-1)^2}{2}-1$$

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x0 = 1
n = 4
```

Asignamos nuestra variable $x$

```python
x=sp.Symbol('x')
```

Calculamos la serie de Mclaurin para nuestros valores iniciales

```python
t=sp.ln(x).series(x,x0,n).removeO()
print(t)
```

```python
t= sp.lambdify(x,t,'numpy')
```

```python
x=np.linspace(-1,3,100)
y=np.log(x)
yt = t(x)
```

Graficamos ambas funciones

```python
plt.figure(figsize=(10,10))
plt.plot(x,y, linestyle="--")
plt.plot(x,yt)
plt.axhline(y=0,color="black", linestyle="--")
plt.axvline(x=0,color="black", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Serie de McLaurin de f(x) = ln(x) orden 4")
plt.grid()
plt.show()
```
